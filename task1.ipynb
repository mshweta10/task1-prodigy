```python
# Continuous - Age
plt.figure(figsize=(8,6)) 
plt.hist(df['age'], bins=17, color='lightgreen', edgecolor='black') 
plt.title('Histogram of Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y') 
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
data['age'].plot(kind='bar', color='skyblue')
plt.title('Category Distribution')
plt.xlabel('Index No.')
plt.ylabel('Age')
plt.grid(axis='y')  # Add gridlines to y-axis
plt.show()
```

```python
#Categorical - Sex
plt.figure(figsize=(8,6))
plt.hist(data['sex'],bins=3,color='lightgreen',edgecolor='black')

plt.xlabel('sex')
plt.ylabel('Frequency')
plt.title('Histogram of Sex Distribution')
plt.show()
```
